```python
import json
import glob
import os
```

```python

FILES = glob.glob("./Knowledge-graph/doc_*/PRAG-QA_*.json")

ALL_QUESTIONS = []

print(f"Found {FILES}.")
```

```python
for file in FILES:
    with open(file, 'r', encoding='utf-8') as f:
        data = json.load(f)
        if isinstance(data, list):
            ALL_QUESTIONS.extend(data)
        elif 'questions' in data:
            ALL_QUESTIONS.extend(data['questions'])
```

```python
os.makedirs("./PRAG/data/scierc/", exist_ok=True)

with open('PRAG/data/scierc/questions.json', 'w', encoding='utf-8') as f:
    json.dump(ALL_QUESTIONS, f, ensure_ascii=False, indent=2)

print(f"Combined {len(ALL_QUESTIONS)} questions in PRAG/src/data/scierc/questions.json")
```
