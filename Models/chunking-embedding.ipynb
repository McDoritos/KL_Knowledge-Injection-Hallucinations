## Importing Libraries

```python
from ollama import embed
import glob
import json
import os
```

## Defining Variables and Function for Chunk Storanging

```python
VECTOR_DB = []

FILES = glob.glob("../Knowledge-graph/doc_*/content_*.txt")

ALL_CHUNKS = []

EMBEDDING_MODEL = "mxbai-embed-large"

def add_chunk_to_vector_db(chunk):
    embedding = embed(model=EMBEDDING_MODEL, input=chunk)['embeddings'][0]
    VECTOR_DB.append((chunk, embedding))
```

## Storing all chunks from content files

```python
for filepath in FILES:
    with open(filepath, "r", encoding="utf-8") as f:
        text = f.read()
        chunks = [line.strip() for line in text.split("\n") if line.strip()]
        ALL_CHUNKS.append(chunks)
        for chunk in chunks:
            add_chunk_to_vector_db(chunk)

print(f"Total chunks processed: {len(VECTOR_DB)}")
```

## Storing on a JSON file

```python
os.makedirs("../Vector-store", exist_ok=True)

with open("../Vector-store/vector_db.json", "w", encoding="utf-8") as f:
    json.dump(
        [{"chunk": c, "embedding": e} for c, e in VECTOR_DB],
        f,
        ensure_ascii=False,
        indent=2
    )
```
